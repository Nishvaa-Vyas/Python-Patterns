```python
#Right Triangle Star Pattern
rows=5
for i in range(1,rows+1):
    for j in range(i):
        print("*",end=" ")
    print()
```

```python
#Inverted Triangle Star Pattern
rows=5
for i in range(rows,0,-1):
    for j in range(i):
        print("*",end=" ")
    print()
```

```python
#Pyramid Star Pattern
rows = 5
for i in range(rows):
    spaces = rows - i - 1
    stars = 2 * i + 1
    print(" " * spaces + "*" * stars)
```
