```python
#Right Triangle Star Pattern
rows=5
for i in range(1,rows+1):
    for j in range(i):
        print("*",end=" ")
    print()
```

```python
#Inverted Triangle Star Pattern
rows=5
for i in range(rows,0,-1):
    for j in range(i):
        print("*",end=" ")
    print()
```

```python
#Pyramid Star Pattern
rows = 5
for i in range(rows):
    spaces = rows - i - 1
    stars = 2 * i + 1
    print(" " * spaces + "*" * stars)
```

```python
#Inverted Pyramid Star Pattern
rows = 5
for i in range(rows):
    spaces = i
    stars = 2 * (rows - i) - 1
    print(" " * spaces + "*" * stars)
```

```python
#Diamond Star Pattern
rows = 5
# upper part
for i in range(1, rows + 1):
    print(" " * (rows - i), end="")
    print("*" * (2*i - 1))
# lower part
for i in range(rows - 1, 0, -1):
    print(" " * (rows - i), end="")
    print("*" * (2*i - 1))
```

```python
#Number Triangle Pattern
rows=5
for i in range(1,rows+1):
    for j in range(1,i+1):
        print(j,end="")
    print()
```

```python
#Same number Triangle Pattern
rows=5
for i in range(1,rows+1):
    for j in range(i):
        print(i,end="")
    print()
```
